```python
# @title merchant_category

from matplotlib import pyplot as plt
import seaborn as sns
data.groupby('merchant_category').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title customer_gender

from matplotlib import pyplot as plt
import seaborn as sns
data.groupby('customer_gender').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title transaction_amount

from matplotlib import pyplot as plt
data['transaction_amount'].plot(kind='hist', bins=20, title='transaction_amount')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title customer_age

from matplotlib import pyplot as plt
data['customer_age'].plot(kind='hist', bins=20, title='customer_age')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
